```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
```

## Tình hình tỉ lệ lao động trên 15 tuổi đã qua đào tạo theo từng ngành trong giai đoạn 2011 - 2020 như thế nào?

### Đọc và khám phá dữ liệu

```python
file_name = 'ti_le_lao_dong.json'
with open(file_name, 'r') as myfile:
    if len(myfile.readlines()) != 0:
        myfile.seek(0)
        data = json.load(myfile)
```

```python
ti_le_lao_dong_da_qua_dao_tao_theo_nganh = pd.DataFrame(data = np.reshape(data['dataset']['value'], (22, 13)), index = data['dataset']['dimension']['Ngành kinh tế']['category']['label'].values(), columns = data['dataset']['dimension']['Năm']['category']['label'].values())
ti_le_lao_dong_da_qua_dao_tao_theo_nganh = ti_le_lao_dong_da_qua_dao_tao_theo_nganh[ti_le_lao_dong_da_qua_dao_tao_theo_nganh.index != 'TỔNG SỐ']
ti_le_lao_dong_da_qua_dao_tao_theo_nganh
```

```python
shape = ti_le_lao_dong_da_qua_dao_tao_theo_nganh.shape
print("Dữ liệu gồm", shape[0], "dòng và", shape[1], "cột")
```

Một số thống kê sơ bộ về bộ dữ liệu trên

```python
ti_le_lao_dong_da_qua_dao_tao_theo_nganh.describe()
```

Nhận xét:
- Theo như thống kê theo 21 ngành nghề, ta có thể thấy tỉ lệ lao động đã qua đào tạo có sự chênh lệch rất lớn giữa các ngành, ngành yêu cầu cao thì có tỉ lệ lao động đã qua đào tạo vượt 80%, ngành yêu cầu thấp thì tỉ lệ chưa tới 10%.
- Đa phần các ngành có tỉ lệ lao động đã qua đào tạo đạt chưa tới 50% và tỉ lệ này vẫn đang tăng theo từng năm, thể hiện rõ nhất ở các ngành có tỉ lệ nằm ở mức giữa
- Các nghề có yêu cầu thấp (tỉ lệ dưới 10%) tăng rất chậm, các ngành đạt tỉ lệ cao trên 80% thì hầu như không có sự thay đổi nhiều sau 10 năm do đã đạt mức bão hòa

### Sự thay đổi sau 10 năm từ 2011 đến 2020 về Top 5 ngành có tỉ lệ lao động đã qua đào tạo cao nhất.

```python
temp_2011 = ti_le_lao_dong_da_qua_dao_tao_theo_nganh['2011'].sort_values(ascending = False).head()
temp_2020 = ti_le_lao_dong_da_qua_dao_tao_theo_nganh['2020'].sort_values(ascending = False).head()
```

```python
fig, ax = plt.subplots(2, 1, figsize = (10, 4), constrained_layout = True)
temp_2011.plot(ax = ax[0], kind='barh', fontsize = 8)
ax[0].set_title("Top 5 ngành có tỉ lệ lao động qua đào tạo cao nhất năm 2011", fontsize = 12)
for index, value in enumerate(temp_2011.values.tolist()):
    ax[0].text(value + 0.5, index, str(round(value, 1)), color = 'blue')

temp_2020.plot(ax = ax[1], kind='barh', fontsize = 8)
ax[1].set_title("Top 5 ngành có tỉ lệ lao động qua đào tạo cao nhất năm 2020", fontsize = 12)
label = list(temp_2020.index)
label[2] = 'Hoạt động Đảng - chính trị'
ax[1].set_yticks(range(5))
ax[1].set_yticklabels(label)
for index, value in enumerate(temp_2020.values.tolist()):
    ax[1].text(value + 0.5, index, str(round(value, 1)), color = 'blue')
```

Nhận xét:
- Có sự thay đổi trong top 5 ngành có tỉ lệ lao động đã qua đào tạo cao nhất, trong đó vẫn có 3 ngành giữ vững vị trí là "Giáo dục và đào tạo", "Y tế và hoạt động trợ giúp xã hội" và "Hoạt động tài chính, ngân hàng và bảo hiểm".
- Các hoạt động chính trị và truyền thông được nâng cao chất lượng lao động và cũng cho thấy mức độ quan tâm, quan trọng của các vấn đề này hiện nay, trong thời đại hội nhập và khoa học công nghệ thì các vấn đề về văn hóa, đạo đức luôn cần được phát huy và giữ gìn.
- Tỉ lệ lao động đã qua đào tạo ở từng thứ hạng năm 2020 đều cao hơn 2011, vì thế mà việc các hoạt động chuyên môn, khoa học bị lọt ra khỏi top 5 không có nghĩa là nó đang thụt lùi hoặc không phát triển, mà chỉ là tốc độ phát triển chậm hơn các ngành trong top 5 mà thôi.

### Top 5 ngành có tốc độ tăng trưởng nhanh nhất về tỉ lệ lao động qua đào tạo trong giai đoạn 2011 - 2020

```python
temp_trend = ti_le_lao_dong_da_qua_dao_tao_theo_nganh['2020'] - ti_le_lao_dong_da_qua_dao_tao_theo_nganh['2011']
temp_trend = temp_trend.sort_values(ascending = False)
```

```python
asc_trend = temp_trend.head()
des_trend = temp_trend.tail()
```

```python
fig, ax = plt.subplots(2, 1, figsize = (10, 4), constrained_layout = True)

asc_trend.plot(ax = ax[0], kind='barh', fontsize = 8)
ax[0].set_title("Top 5 ngành có tỉ lệ lao động qua đào tạo tăng nhiều nhất", fontsize = 12)
label = list(asc_trend.index)
label[3] = 'Buôn bán và sửa chữa xe'
label[2] = 'Hoạt động Đảng - chính trị'
ax[0].set_yticks(range(5))
ax[0].set_yticklabels(label)
for index, value in enumerate(asc_trend.values.tolist()):
    ax[0].text(value + 0.2, index, str(round(value, 1)), color = 'blue')
    
des_trend.plot(ax = ax[1], kind='barh', fontsize = 8)
ax[1].set_title("Top 5 ngành có tỉ lệ lao động qua đào tạo tăng ít nhất", fontsize = 12)
label = list(des_trend.index)
label[3] = 'Giúp việc, sản xuất sản phẩm và dịch vụ hộ gia đình'
ax[1].set_yticks(range(5))
ax[1].set_yticklabels(label)
for index, value in enumerate(des_trend.values.tolist()):
    if value > 0:
        ax[1].text(value + 0.02, index, str(round(value, 1)), color = 'blue')
    else:
        ax[1].text(value - 0.12, index, str(round(value, 1)), color = 'blue')
```

Nhận xét:
- Trong top 5 ngành có tốc độ tăng trưởng ít nhất vầ tỉ lệ lao động đã qua đào tạo, có 2 ngành "Hoạt động của tổ chức và cơ quan quốc tế" và "Giáo dục và đào tạo" nằm trong top 5 của năm 2011. Như vậy ta có thể thấy 2 ngành này đã đạt mức độ bão hòa, và sẽ tăng trưởng rất chậm hoặc biến động xung quanh chỉ số hiện tại.
- "Nông nghiệp, lâm nghiệp và thủy sản" với tỉ lệ 2.7% và "Xây dựng" 11.7% vào năm 2011 nằm trong top những ngành có tỉ lệ lao động đã qua đào tạo thấp nhất thì với mức tăng trưởng ít như vậy thật ra cũng đã là nhiều so với con số vốn có lúc ban đầu.

## Tình hình số lượng lao động chia theo nghề nghiệp ở nước ta như thế nào?

### Đọc và khám phá dữ liệu

```python
file_name = 'lao_dong_theo_nghe.json'
with open(file_name, 'r') as myfile:
    if len(myfile.readlines()) != 0:
        myfile.seek(0)
        data = json.load(myfile)
```

```python
print(data['dataset']['dimension']['Nghề nghiệp']['category']['label'])
```

```python
phan_bo_lao_dong_theo_nghe_nghiep = pd.DataFrame(data = np.reshape(data['dataset']['value'], (11, 13)), index = data['dataset']['dimension']['Nghề nghiệp']['category']['label'].values(), columns = data['dataset']['dimension']['Năm']['category']['label'].values())
phan_bo_lao_dong_theo_nghe_nghiep = phan_bo_lao_dong_theo_nghe_nghiep[phan_bo_lao_dong_theo_nghe_nghiep.index != 'TỔNG SỐ']
phan_bo_lao_dong_theo_nghe_nghiep
```

```python
shape = phan_bo_lao_dong_theo_nghe_nghiep.shape
print("Dữ liệu gồm", shape[0], "dòng và", shape[1], "cột")
```

```python
phan_bo_lao_dong_theo_nghe_nghiep = phan_bo_lao_dong_theo_nghe_nghiep.astype('float')
phan_bo_lao_dong_theo_nghe_nghiep.describe()
```

Nhận xét:
- Có sự cách biệt lớn về số lượng lao động giữa các ngành nghề với nghề có số lượng lớn lên đến hơn 18000, nghề có ít lao động chỉ khoảng 100 người.
- Nghề có ít lao động có sự biến động ít giữa các năm và trong cả giai đoạn 2011 - 2020, nghề có số lao động cao nhất thì giảm dần dù cho tổng số lượng lao động đang ngày càng tăng, như vậy ta có thể kết luận lao động của nhóm này đang chuyển dịch sang những nhóm khác.

### Cơ cấu lao động theo nghề nghiệp trong 2 năm 2011 và 2020

```python
fig, ax = plt.subplots(1, 2, figsize = (8, 8))

ax[0].pie(phan_bo_lao_dong_theo_nghe_nghiep['2011'], autopct='%.2f%%', labels = None, pctdistance=1.2)
ax[0].set_title('2011', y = -0.05)

ax[1].pie(phan_bo_lao_dong_theo_nghe_nghiep['2020'], autopct='%.2f%%', labels = None, pctdistance=1.2)
ax[1].set_title('2020', y = -0.05)

fig.legend(labels = phan_bo_lao_dong_theo_nghe_nghiep.index, loc="right", bbox_to_anchor=(1.4, 0.5))
fig.suptitle('Biểu đồ tỉ trọng theo vị trí nghề nghiệp', y = 0.72)
```

Nhận xét:
- "Nhà lãnh đạo" có số lượng lao động ổn định.
- Giảm ở nhóm "Nghề giản đơn", "Nghề nông, lâm, ngư nghiệp"
- Tăng ở nhóm "Thợ lắp ráp và vận hành máy móc, thiết bị", "Chuyên môn kỹ thuật bậc cao", "Dịch vụ, bảo vệ, bán hàng" và "Thợ thủ công".
- Cơ cấu lao động ở các nhóm yêu cầu ít tri thức, kỹ thuật hoặc kiến thức chuyên môn đang dần chuyển dịch sang nhóm có chuyên môn cao hơn. Các nhóm ngành này do có sự tác động của công nghệ, thiết bị, máy móc hiện đại thay thế cho sức lao động của con người nên nhu cầu cũng giảm đi.
- Cơ cấu lao động đang dần chuyển dịch sang các ngành yêu cầu chuyên môn cao hoặc không thể thực hiện bằng máy móc (thợ thủ công).

### Tốc độ tăng trưởng theo từng nghề trong giai đoạn 2011 - 2020

```python
temp_trend = (phan_bo_lao_dong_theo_nghe_nghiep['2020'] - phan_bo_lao_dong_theo_nghe_nghiep['2011']) / phan_bo_lao_dong_theo_nghe_nghiep['2011'] * 100
temp_trend = temp_trend.sort_values(ascending = False)
asc_trend = temp_trend.head()
des_trend = temp_trend.tail()
```

```python
fig, ax = plt.subplots(2, 1, figsize = (10, 4), constrained_layout = True)

asc_trend.plot(ax = ax[0], kind='barh', fontsize = 10)
ax[0].set_title("Top 5 nghề có tốc độ tăng trưởng cao nhất về số lượng lao động", fontsize = 12)
for index, value in enumerate(asc_trend.values.tolist()):
    ax[0].text(value + 1, index, str(round(value, 1)), color = 'blue')
    
des_trend.plot(ax = ax[1], kind='barh', fontsize = 10)
ax[1].set_title("Top 5 nghề có tốc độ tăng trưởng thấp nhất về số lượng lao động", fontsize = 12)
for index, value in enumerate(des_trend.values.tolist()):
    if value > 0:
        ax[1].text(value + 0.5, index, str(round(value, 1)), color = 'blue')
    else:
        ax[1].text(value - 2.8, index, str(round(value, 1)), color = 'blue')
```

Nhận xét:
- Nhu cầu lao động tăng cao ở các ngành yêu cầu ký thuật, kiến thức chuyên môn cao hoặc máy móc không thể thay thế con người, nổi bật nhất là nhóm "Thợ lắp ráp và vận hành" tăng gấp đôi về số lượng trong giai đoạn này.
- Các nhóm ngành bị thay thế hoặc được trợ giúp cao từ quá trình công nghiệp hóa, hiện đại hóa có sự giảm mạnh điển hình là "Nông, lâm, ngư nghiệp" giảm một nửa và "Nghề giản đơn" giảm 1/10 trên tổng số dù cho tổng số lượng người lao động trong nền kinh tế vẫn đang ngày một tăng.
