```python
import pandas as pd
import numpy as np
import json
```

```python
file_name = 'data/V02.53.json'
with open(file_name, 'r') as myfile:
    if len(myfile.readlines()) != 0:
        myfile.seek(0)
        data = json.load(myfile)
```

```python
value = data['dataset']['value']
value = np.array(value)
```

```python
new_value = []
for i in range(0, int(len(value) / 8)):
    new_value.append(value[8 * i : 8 * i + 4])
new_value = np.array(new_value)
```

```python
# ti_le_that_nghiep = pd.DataFrame(data = new_value[7:14], index = data['dataset']['dimension']['Vùng']['category']['label'].values(), columns = data['dataset']['dimension']['Nhóm tuổi']['category']['label'].values())
ti_le_that_nghiep = pd.DataFrame(data = new_value, index = list(data['dataset']['dimension']['Vùng']['category']['label'].values()) * 12, columns = data['dataset']['dimension']['Nhóm tuổi']['category']['label'].values())
temp = list(data['dataset']['dimension']['Năm']['category']['label'].values())
temp = np.repeat(temp, 7)
ti_le_that_nghiep['Năm'] = temp
ti_le_that_nghiep
```
